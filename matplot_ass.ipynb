```python
#read the csv file
#creating data frame from a csv file removing the total_unit and total_profit column

headers= ["month_number","facecream","facewash", "toothpaste",	"bathingsoap",	"shampoo",	"moisturizer" ]
df_a = pd.read_csv("/content/drive/MyDrive/Pasndas_intro/company_sales_data.csv", usecols= headers)

#check the file
df_a.head()
```

```python
#plotting the graph of sales/quantity of product per month

plt.rcParams["figure.figsize"] = (10,6)

plt.plot(df_a.set_index('month_number'),linestyle = "--", marker = "s")

#the code above plot all product sales in a month seting month_number as x-values
#with line style and marker


#labeling the x-axis and y-axis
plt.xlabel("Months")
plt.ylabel("Quantity")

#giving the chart a title and also naming each line using the legend fucntion
#the loc specify the corner of axes to put the legend
plt.title("Quantity of product sales per month", loc= "left")
plt.legend(["facecream","facewash", "toothpaste",	"bathingsoap",	"shampoo",	"moisturizer"], loc="upper left")

plt.show()
```

```python
#finding the sum of sales of each product dropping the month_number, total_unit and total_profit

#reading the csv file
dfp = pd.read_csv("/content/drive/MyDrive/Pasndas_intro/company_sales_data.csv")


#droppin the month_number, total_unit and total_profit
dfp.drop(["month_number", "total_units",	"total_profit"], axis= 1, inplace = True)

#summing each product and renaming the column
df3= dfp.sum(axis= 0).rename_axis("product").reset_index(name="sum")

df3.head()
```

```python
from pandas.core.frame import DataFrame
#crepresenting the anual sales of each product in pie chart
#using the process from previous cose to get the summation.

# renaming the ararqhuement needed
x= df3["sum"]
y= df3["product"]


#plotting the pie chart
#the autopct argurement add percentage level to the chart

plt.pie(x , labels= y, shadow= True, autopct="%1.2f%%")
# plt.axis('equal')

plt.title("Total sales for each product for the year")
plt.show()

filename= "product pie-chart"
plt.savefig(filename+'.png')
```
